#2차원 배열

```python
a = np.arange(30).reshape(5,6) # 0부터 29까지의 숫자로 5행 6열 2차원 배열 생성
a
```

```python
a[3] # 인덱스 3에 해당하는 네 번째 행 전체를 선택
```

```python
a[-1] # 마지막 행 전체를 선택
```

```python
a[-2,:] # 뒤에서 두 번째 행 전체를 선택 (':'은 해당 축의 모든 요소를 의미)
```

```python
a[-2,] # 뒤에서 두 번째 행 전체를 선택 (a[-2,:]와 동일한 결과)
```

```python
a[2:4] # 인덱스 2부터 3까지의 행(세 번째 행, 네 번째 행)을 선택
```

```python
a[1:5:2] # 인덱스 1부터 4까지의 행 중 2단계씩 건너뛰어 선택 (두 번째 행, 네 번째 행)
```

```python
a[,1]
```

```python
a[:,1] # 모든 행에서 인덱스 1에 해당하는 두 번째 열을 선택
```

```python
a[:,1:6] # 모든 행에서 인덱스 1부터 5까지의 열을 선택 (두 번째 열부터 마지막 열까지)
```

```python
a = np.arange(30).reshape(5,6) # 0부터 29까지의 숫자로 5행 6열 2차원 배열 생성
a
```

```python
a[:,::2] # 모든 행에서 짝수 번째 열만 선택 (0, 2, 4열)
```

```python
a[2,4] # 인덱스 (2,4)에 해당하는 단일 요소 선택 (세 번째 행, 다섯 번째 열)
```

```python
a[2][4] # 행을 먼저 선택한 후 해당 행에서 열을 선택 (a[2,4]와 동일)
```

```python
a[1:3,2:5] # 행 인덱스 1부터 2까지(두 번째, 세 번째 행)와 열 인덱스 2부터 4까지(세 번째, 네 번째, 다섯 번째 열)를 선택
```

```python
a[:,:] # 배열 전체를 선택 (모든 행, 모든 열)
```

```python
a[:,] # 배열 전체를 선택 (a[:,:]와 동일)
```

```python
a[:] # 배열 전체를 선택 (a[:,:]와 동일)
```

```python
a[::-1,::-1] # 배열을 행과 열 모두 역순으로 뒤집기
```

```python
# 마지막 행과 열을 제외한 배열 반환
a[:-1,:-1] # 마지막 행과 마지막 열을 제외한 모든 요소를 선택
```

#3차원 배열

```python
a=np.arange(24).reshape(2,3,4) # 0부터 23까지의 숫자로 2층 3행 4열 3차원 배열 생성
a
```

```python
a[1] # 인덱스 1에 해당하는 두 번째 '층'(axis=0) 전체를 선택
```

```python
a[:,1] # 모든 '층'에서 인덱스 1에 해당하는 두 번째 '행'(axis=1) 전체를 선택
```

```python
a[:,1,1] # 모든 '층'에서 인덱스 1의 '행', 그리고 인덱스 1의 '열'(axis=2)에 해당하는 요소들을 선택
```

```python
a[:,:,1] # 모든 '층'과 모든 '행'에서 인덱스 1에 해당하는 두 번째 '열' 전체를 선택
```

```python
a[..., 1] # '...'은 생략된 모든 차원을 의미합니다. 마지막 차원의 인덱스 1을 선택 (a[:,:,1]과 동일)
```

```python
a[1, ..., 1] # 첫 번째 차원의 인덱스 1(두 번째 층)에서, 생략된 중간 차원들을 거쳐 마지막 차원의 인덱스 1을 선택
```

```python
a[0, ...] # 첫 번째 차원의 인덱스 0(첫 번째 층)을 선택하고 나머지 차원은 모두 포함 (a[0,:,:] 또는 a[0]과 동일)
```

```python
a[0] # 첫 번째 차원의 인덱스 0(첫 번째 층) 전체를 선택 (a[0,...]과 동일)
```
